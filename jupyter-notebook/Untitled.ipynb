```python
import picamera
import PIL
from PIL import Image
```

```python
camera = picamera.PiCamera()
```

```python
res = camera.resolution
```

```python
SINGLE_FRAME_SIZE_YUV = res[0] * res[1] * 3//2
```

```python
YUV420P_Y_PLANE_SIZE = res[0] * res[1]
YUV420P_U_PLANE_SIZE = YUV420P_Y_PLANE_SIZE //4
YUV420P_V_PLANE_SIZE = YUV420P_U_PLANE_SIZE
```

```python
stream = picamera.PiCameraCircularIO(camera, size=SINGLE_FRAME_SIZE_YUV)
```

```python
camera.framerate = 8
camera.brightness = 60
camera.start_recording(stream, format = 'yuv')
```

```python
cam_output = stream.getvalue()
if len(cam_output) != SINGLE_FRAME_SIZE_YUV:
    assert False
y_plane = cam_output[0:YUV420P_Y_PLANE_SIZE]
u_plane = cam_output[YUV420P_Y_PLANE_SIZE:
                     YUV420P_Y_PLANE_SIZE+YUV420P_U_PLANE_SIZE]
v_plane = cam_output[YUV420P_Y_PLANE_SIZE+YUV420P_U_PLANE_SIZE:
                    SINGLE_FRAME_SIZE_YUV]

img = Image.fromarray(cam_output)
img.show()
```
